```python
#
#     Exploring some questions:
#
#
#
# Question:
#     
#       How is the victory status divided by each side?
#
#
import csv
import re
import pandas as pd

df = pd.read_csv('games.csv')
df_pivot = pd.pivot_table(df, index=['winner'],aggfunc=len, columns=['victory_status'], fill_value=0, values='id')
df_pivot
```

**Victory Status by Side**

*Process*

This piece of isolated data was obtained through creating a pivot table using the winning side as the index and the victory_status as the column value. A 'len' aggregating function was also applied. The dataframe was directly read from the source 'games.csv' dataset.

A pie chart was then created to display this data.

```python
pieChart1 = df_pivot.plot.pie(title='Victory Status vs. Winner', subplots=True, figsize=(30, 6), colormap='bone',fontsize=20)
pieChart1
```

*Findings*

An observation of this particular table of data and related charts indicates that the white side obtains more checkmates and resigns more often. 

Possible explanations for this are that in the official chess rules, the white player goes first. Thus, more likely to obtain a victory or able to resign from the game before the black side could make their first move.

The black side obtains very slightly more outoftime end games but not by a significant amount to be statistically significant. 

```python
#
# Question:
#     
#       Are some types of victories more common in rated vs unrated games?
#
#
df['rated'] = df.apply((lambda p: str(p['rated']).lower()), axis=1)
df['rated'] = df.apply((lambda p: "rated" if p['rated'] == 'true' else "unrated"), axis=1)
df_pivot = pd.pivot_table(df, index=['rated'],aggfunc=len, columns=['victory_status'], fill_value=0, values='id')
df_pivot['total'] = df_pivot['draw'] + df_pivot['mate'] + df_pivot['outoftime'] + df_pivot['resign']
df_pivot['draw'] = df_pivot['draw'] / df_pivot['total'] * 100
df_pivot['mate'] = df_pivot['mate'] / df_pivot['total'] * 100
df_pivot['outoftime'] = df_pivot['outoftime'] / df_pivot['total'] * 100
df_pivot['resign'] = df_pivot['resign'] / df_pivot['total'] * 100

df_pivot

```

**Comparing Victories in Rated vs Unrated Games**

*Process*

The 'rated' data column was extracted from the application of a lambda function on the dataframe that represents the dataset. The text was cleaned to make all string values lowercase. Through this process, the data was organized between chess games that were rated and unrated by another lambda function.

From this data organization, a pivot table was created by using the 'rated' data as the index and 'victory_status' as the columns. The pivot table data was then assembled for the 'total' header by simply concatenating several individual data pieces. This was then used to complete the pivot table with the remaining columns of 'draw', 'mate', 'outoftime', and 'resign'

A pie chart was selected as displaying the data in the most effective way.

```python
pieChart1 = df_pivot.plot.pie(title='Victory Status vs. Game Status', subplots=True, figsize=(20, 6),fontsize=20)
pieChart1
```

*Findings*

From the various pie charts, it can be observed that rated games and unrated games are quite similar in terms of the small magnitude of their differences. But there is one notable exception.

Unrated games appear to have far more 'draw' game endings than rated games. 

While rated games have slightly more 'checkmates' and 'outoftime' game endings than unrated games. But these differences are slight compared to the 'draw' game ending.

Rated and unrated games appear to have an almost equal amounts of 'resigned' game endings. 

A possible explanation for the comparison of rated and unrated games are that unrated games might have more unexperienced players who are more prone to getting into the position of a draw. But there isn't a clear definitive reason for this occurance.

```python
#
# Question:
#     
#       Which side initiates more draws?
#
#
df_draws = df.filter(items=['turns', 'winner'])
df_draws['last move'] = df.apply( (lambda p: "black" if int(p['turns']) % 2 == 0 else "white" ), axis=1)
df_pivot = pd.pivot_table(df_draws, index=['last move'],aggfunc=len, columns=['winner'], fill_value=0, values='turns')
df_pivot
```

**Comparing The White & Black Side in Game Draws**

*Process*

A new dataframe was created (df_draws) that was assembled through filtering through the dataframe that represents the dataset. The filter command used 'turns' and 'winner' as the items. 

From this newly created dataframe, the 'last move' column was isolated for and a lambda function was applied. This lambda function seperated the last moves by the black and white side (i.e. which player ended the game) 

A pivot table was created from this dataframe using the 'last move' data as the index and the 'winner' data as the columns. The 'len' aggregation function was applied.

```python
pieChart1 = df_pivot.plot.pie(title='Last move into draw for both sides', y='draw', figsize=(20, 6),fontsize=20)
pieChart1
```

*Findings*

This singular pie chart shows that the white side causes slightly more 'draw' game endings. 

This trend correlates with the previous trends found in this project with the white side having more victories but also more resignations than the black side. 

A possible explanation for those previous observations was that because the white side goes first, game ending events are more likely to be caused by the white side. In this instance, it's possible that this explanation could be the reason for this trend.

```python
#
# Question:
#     
#       Are there differences between the victory status and the winners rating?
#
#

def ratingToStr(x):
    x = int(x)
    if x < 1200:
        return "very low"
    elif x < 1500:
        return "low"
    elif x < 2000:
        return "average"
    elif x < 2300:
        return "high"
    elif x < 3000:
        return "very high"

ord_skill = ["very high", "high", "average", "low", "very low"]
df["Winner rating"] = df.apply((lambda p: ratingToStr(p['white_rating'] if p['winner'] == "white" else p['black_rating'])), axis=1)
df_pivot = pd.pivot_table(df, index=['Winner rating'],aggfunc=len, columns=['victory_status'], fill_value=0, values='turns')
df_pivot = df_pivot.reindex(ord_skill)
df_pivot['total'] = df_pivot['draw'] + df_pivot['mate'] + df_pivot['outoftime'] + df_pivot['resign']
df_pivot['draw (%)'] = df_pivot['draw'] / df_pivot['total'] * 100
df_pivot['mate (%)'] = df_pivot['mate'] / df_pivot['total'] * 100
df_pivot['outoftime (%)'] = df_pivot['outoftime'] / df_pivot['total'] * 100
df_pivot['resign (%)'] = df_pivot['resign'] / df_pivot['total'] * 100
df_pivot 
```

**Skill Rating Compared to Victory Status**

*Process*

A function is created that is named 'ratingToStr' that takes in a singular numeric skill rating and converts it to a string describing the skill rating in question. These categories are very low, low, average, high, and very high.

This function is then applied and the data is saved to the 'winner rating' column. A pivot table is created using similar methods described in previous pivot table constructions in this project. This pivot table includes raw numeric data for each category as well as percentages.

*Findings*

The data, when split by the winner's skill rating, helps give characteristics to how chess games at various skill levels are played.

Games where the winner has a 'very low' skill rating are characterized by more frequent checkmates and an almost equal amount of resignations.

Games where the winner has a 'very high' skill rating are characterized by frequent resignations and sparse amounts of checkmates.

The relationship between skill rating and the percentage occurance of a checkmate is a negative relationship. Higher skilled players are less likely to checkmate against each other. On the opposite end, the relationship between skill rating and the percentage occurance of a resignation is a positive relatiosnhip. Higher skilled players are more likely to simply resign from the game.

For other characteristics, it appears draws and outoftime game endings are more common on the extreme ends of skill levels (very low or very high). 

```python
#
# Question:
#     
#       Are there correlations between the victory status and skill disparity of a match?
#
#
def skillDiff(a, b):
    x = abs(int(a) - int(b))
    if x < 100:
        return "very low"
    elif x < 200:
        return "low"
    elif x < 300:
        return "average"
    elif x < 500:
        return "high"
    elif x < 3000:
        return "very high"
    
df["Skill Difference"] = df.apply((lambda p: skillDiff(p['white_rating'], p['black_rating'])), axis=1)
df_pivot = pd.pivot_table(df, index=["Skill Difference"],aggfunc=len, columns=['victory_status'], fill_value=0, values='turns')
df_pivot = df_pivot.reindex(ord_skill)
df_pivot['total'] = df_pivot['draw'] + df_pivot['mate'] + df_pivot['outoftime'] + df_pivot['resign']
df_pivot['draw (%)'] = df_pivot['draw'] / df_pivot['total'] * 100
df_pivot['mate (%)'] = df_pivot['mate'] / df_pivot['total'] * 100
df_pivot['outoftime (%)'] = df_pivot['outoftime'] / df_pivot['total'] * 100
df_pivot['resign (%)'] = df_pivot['resign'] / df_pivot['total'] * 100
df_pivot 
```

**Skill Rating Difference Compared to Victory Status**

*Process*

A function is created that is named 'skillDiff' that takes in the white side skill rating and the black skill rating, subtracting both to get the difference after applying the absolute value function. This numeric difference is then converted to a string representation of either very low, low, average, high, and very high. Lambda notation is used to create a new data column of skill differences.

A pivot table is then assembled in a similar manner to the construction of previous pivot tables found in this project.

*Findings*

It seems, from the data above, that the skill disparity of two players does not cause significant differences in most game endings.

It appears that games with a lower skill difference tend to result in more draws. A negative relationship is observed through this trend. 

It also appears that games with a higher skill difference tend to result in more checkmates. A positive relationship is observed through this trend.

The trend for 'outoftime' game endings is not as strong but it tends to appear more likely in games with a lower skill disparity. The trend for resignations is not as strong as well but, like with 'outoftime' game endings, tends to occur more often with games of a lower skill disparity.

Overall, these differences are only slight as the statistics tend to show that these games are quite similar, despite skill disparity. There isn't any large difference between the various skill disparities.
